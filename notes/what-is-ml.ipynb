## What is Machine Learning and Its Types

### Definition of Machine Learning

Machine Learning (ML) is like teaching computers to learn from examples and improve on their own. It involves:

1. Collecting data
2. Creating a mathematical model based on that data
3. Using the model to make predictions or decisions

### Types of Machine Learning

#### 1. Supervised Learning

In supervised learning, we give the computer examples with correct answers. It's like a teacher guiding a student.

- **Example**: Predicting house prices
  - Input: House size, location, number of rooms
  - Output: Price

The computer learns to connect the inputs to the correct outputs.

#### 2. Unsupervised Learning

In unsupervised learning, we give the computer data without any answers. It tries to find patterns on its own.

- **Example**: Grouping customers
  - Input: Customer shopping habits
  - Output: Groups of similar customers

The computer finds similarities and differences without being told what to look for.

#### 3. Semi-Supervised Learning

This is a mix of supervised and unsupervised learning. We give the computer some data with answers and some without.

- **Example**: Classifying web pages
  - Some web pages are labeled (sports, news, etc.)
  - Many are unlabeled

The computer uses both labeled and unlabeled data to learn.

#### 4. Reinforcement Learning

In reinforcement learning, the computer learns by trying things out and getting rewards or punishments.

- **Example**: Teaching a computer to play a video game
  - Action: Move left, right, jump, etc.
  - Reward: Points scored in the game

The computer learns which actions lead to the highest rewards.

### Conclusions

Machine Learning helps computers solve problems by learning from data. Different types of ML are suited for different kinds of problems. As you continue learning, you'll see how these methods are used in real-world applications.

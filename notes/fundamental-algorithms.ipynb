### Learning from data

As we’ve learned, in machine learning, we typically create models that predict or classify input values. We can think of a model as a function or an equation that describes the relationship between the input and output.

For example:
$$ y = 2x + 3 $$
This is a simple equation where  y  is the output and  x  is the input.

More complex models might have multiple inputs:
$$ y = 2x_1 + 3x_2 + 4 $$

The general form of the fucntion is:
$$ y = mx + c $$

You are familar with this form of the function from your algebra classes.  Here, 

$y$ is the output, 
$x$ is the input, 
$m$ is the weight, 
$c$ is the intercept.

In machine learning we write typically write this function slightly differently:
$$ y = wx + b $$

or in case of multiple inputs:
$$ y = w_1x_1 + w_2x_2 + ... + w_nx_n + b $$

Here, 
$w$ is the weight, 
$x$ is the input, 
$b$ is the intercept.

Our goal in machine learning is to learn the weights (like 2 and 3) and the intercept (like 4) from data, allowing us to make predictions.

### The Three Key Parts of a Machine Learning Algorithm

Before diving deeper into machine learning algorithms like Linear Regression and Logistic Regression, it’s important to understand that all machine learning algorithms share three key components:
1. The Function (or Model): This is the equation or formula as depicted above, that represents the relationship between the input data and the output predictions. The function helps us make predictions based on input data.
2.	The Loss Function: This tells us how well the model is doing at making predictions. The loss function measures the difference between the model’s predictions and the actual data. The goal is to minimize this loss as much as possible.
3.	The Optimization Function: This is the process that helps us improve the model by adjusting its parameters to minimize the loss. The optimization function systematically updates the model to make better predictions.

For example, you could use the function below. Here,  x  is your position on the hill, and  y  is the height. The slope  m  and the intercept  c  describe the shape of the hill.
$$ y = mx + c $$


2.	The Loss Function: The loss function tells you how far you are from the bottom of the hill (the optimal point). If you’re far from the bottom, your “loss” is high. If you’re close to the bottom, the loss is low. In this example, the loss is the height of the hill: the higher up you are, the greater the loss.
3.	The Optimization Function: The optimization function is like your strategy for getting down the hill. You might use a strategy like “follow the steepest path downhill.” This is similar to Gradient Descent in machine learning, where the model looks for the steepest direction to reduce the loss. With each step, you adjust your position (the model’s parameters) to get closer to the lowest point (minimize the loss).

### Loss Functions

Whenever we build a model, we need to evaluate how well it works. This is done using loss functions. A loss function tells us how far off our model's predictions are from the actual results. The lower the loss, the better the model is.

The most common loss function is the Mean Squared Error (MSE).

### Mean Squared Error (MSE)

For regression problems, where we predict continuous values (like house prices or test scores), we use Mean Squared Error (MSE). MSE measures the average squared difference between the predicted values ($\hat{y}$) and the actual values ($y$):

$$ \text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $$

Here's an example using actual values to make this more concrete:

- Suppose we have data points: $x = [1, 2, 3]$ and corresponding actual values $y = [2, 4, 6]$.
- Let's assume our model's predictions are $\hat{y} = [1.8, 4.2, 5.9]$.

To compute the MSE:

$$ \text{MSE} = \frac{1}{3} \left[(2 - 1.8)^2 + (4 - 4.2)^2 + (6 - 5.9)^2 \right] $$

Breaking it down:

- $(2 - 1.8)^2 = 0.04$
- $(4 - 4.2)^2 = 0.04$
- $(6 - 5.9)^2 = 0.01$

Summing them and averaging:
$$ \text{MSE} = \frac{1}{3} (0.04 + 0.04 + 0.01) = 0.03 $$

This tells us that, on average, our model's predictions are off by about 0.03 units from the actual values.

### Binary Cross-Entropy

For classification problems, where the goal is to predict categories (like “pass” or “fail”), we use a different loss function called Binary Cross-Entropy. This measures how well the model is predicting probabilities for binary outcomes (0 or 1).

The formula is:

$$ L = - \frac{1}{n} \sum_{i=1}^{n} \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right] $$

Here’s an example using actual values:

 - Suppose we have binary outcomes  y = [0, 1, 1]  and our model’s predicted probabilities are  \hat{y} = [0.2, 0.8, 0.9] .

To compute the Binary Cross-Entropy loss:

$$ L = - \frac{1}{3} \left[ 0 \cdot \log(0.2) + 1 \cdot \log(0.8) + 1 \cdot \log(0.9) \right] $$

Breaking it down:
 - For the first data point:  0 \cdot \log(0.2) = 0 
 - For the second data point:  \log(0.8) = -0.223 
 - For the third data point:  \log(0.9) = -0.105 

Summing and averaging:

$$ L = - \frac{1}{3} \left[ 0 + (-0.223) + (-0.105) \right] = 0.109 $$

This value represents the average error in predicting the binary class probabilities. The closer the loss is to 0, the better the model is performing.

### Linear Regression

In Linear Regression, we try to fit a straight line to the data points, and we try to minimize the Mean Squared Error (MSE) between the predicted values and actual values.

The equation of the line is:
$$ y = mx + c $$

Where:

- $y$ is the predicted output,
- $x$ is the input,
- $m$ is the slope,
- $c$ is the intercept.

#### Ordinary Least Squares (OLS) Method

The Ordinary Least Squares (OLS) method is a way to find the best-fit line by minimizing the MSE directly. It calculates the slope $m$ and intercept $c$ by solving for the values that make the squared differences between the actual values and the predicted values as small as possible.

In OLS, we derive formulas for $m$ and $c$:

$$ m = \frac{\sum{(x_i - \bar{x})(y_i - \bar{y})}}{\sum{(x_i - \bar{x})^2}} $$
$$ c = \bar{y} - m \cdot \bar{x} $$

Where $\bar{x}$ and $\bar{y}$ are the means of the $x$ and $y$ values.

#### Linear Regression Example (OLS) with MSE

Now, let's walk through the steps of Linear Regression with OLS in Python. First, we will plot the data points without the line, and then we will plot the line after calculating $m$ and $c$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 2*x + 1
```

```python
# Example data (x, y)
x = np.array([1, 2, 3, 4, 5])
y = f(x)  # y = 2x + 1 + some noise
```

```python
# Plot the data points
plt.scatter(x, y, color='blue', label='Data points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data Points for Linear Regression')
plt.legend()
plt.show()
```

```python
# Use Ordinary Least Squares to compute m and c
x_mean = np.mean(x)
y_mean = np.mean(y)
m = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean)**2)
c = y_mean - m * x_mean

print(f"OLS slope: {m}, intercept: {c}")
```

```python
# Plot the data points with the fitted line
x_vals = np.linspace(0, 6, 100)
y_vals = m * x_vals + c
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x_vals, y_vals, color='red', label=f'y = {m:.2f}x + {c:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fitted Line using OLS')
plt.legend()
plt.show()
```

```python

# Calculate MSE with actual values
y_pred = m * x + c
mse = np.mean((y - y_pred)**2)
print(f"Mean Squared Error (MSE): {mse}")
```

#### Explanation:

- We first plot the data points without the line.
- Using Ordinary Least Squares (OLS), we calculate the slope $m$ and intercept $c$.
- After calculating $m$ and $c$, we plot the fitted line along with the data points.
- We then compute the Mean Squared Error (MSE) to see how well the line fits the data.

### Logistic Regression (Without Gradient Descent)

In Logistic Regression, we predict probabilities for binary outcomes using the logistic (sigmoid) function:

$$ \hat{y} = \frac{1}{1 + e^{-z}} $$

Where $z = mx + c$.

We typically solve for $m$ and $c$ using Maximum Likelihood Estimation (MLE), but let's use a simplified example where we already know the values for $m$ and $c$.

### Logistic Regression Example (Without Gradient Descent) with Binary Cross-Entropy

Just like with Linear Regression, we will first plot the data points without the logistic regression curve and then add the curve after calculating the values for $m$ and $c$.

```python
# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Example binary data
x = np.array([1, 2, 3, 4, 5])
y = np.array([0, 0, 0, 1, 1])  # Binary outcomes
```

```python
# Plot the data points without the curve
plt.scatter(x, y, color='blue', label='Data points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data Points for Logistic Regression')
plt.legend()
plt.show()
```

```python
# Use a simple logistic model with predefined weights
m, c = 1, -3  # Assume we found these values
z = m * x + c
y_pred = sigmoid(z)
```

```python
# Plot logistic regression curve with data points
x_vals = np.linspace(-10, 10, 100)
y_vals = sigmoid(m * x_vals + c)
plt.plot(x_vals, y_vals, color='red', label='Logistic Regression Curve')
plt.scatter(x, y, color='blue', label='Data points')
plt.xlabel('x')
plt.ylabel('Probability (y)')
plt.title('Logistic Regression with Predefined Weights')
plt.legend()
plt.show()
```

```python
# Calculate Binary Cross-Entropy Loss with actual values
bce_loss = -np.mean(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))
print(f"Binary Cross-Entropy Loss: {bce_loss}")
```

#### Explanation

- **Data Points**: We start by plotting the binary data points. In this example, $x$ represents some input (e.g., hours studied) and $y$ represents the binary outcome (e.g., pass = 1, fail = 0).

- **Sigmoid Function**: We use the sigmoid function to predict the probability that $y$ equals 1. The sigmoid function squashes any input $z = mx + c$ into a probability between 0 and 1.

- **Predefined Weights**: Instead of using optimization techniques (like Gradient Descent), we assume predefined values for $m$ and $c$ (in this case, $m = 1$ and $c = -3$).

- **Plotting**: We then plot the logistic regression curve on top of the data points. The curve shows the predicted probability for $y$ as $x$ changes.

- **Binary Cross-Entropy Loss**: Finally, we compute the Binary Cross-Entropy Loss, which tells us how well the model is performing by comparing the predicted probabilities with the actual binary outcomes.

In real-world applications, we typically do not assume the values for $m$ and $c$. Instead, we use optimization methods like Gradient Descent to find the best values that minimize the loss function, making the model better at predicting outcomes.

### Introducing Gradient Descent

Now that we've explored Linear and Logistic Regression using traditional methods (like Ordinary Least Squares (OLS) for Linear Regression), let's introduce a more general method to optimize our models: Gradient Descent.

#### What is Gradient Descent?

Gradient Descent is an optimization algorithm used to minimize a loss function by iteratively adjusting the model's parameters (like $m$ and $c$ in Linear Regression). It's like a hiker trying to find the lowest point on a hill. At each step, the hiker looks at the steepness of the slope (the gradient) and takes a step downhill. Over time, the hiker gets closer and closer to the bottom of the hill.

For machine learning models, this means adjusting the parameters (weights) in small steps, moving in the direction that reduces the error (loss).

##### The Update Rule

In Gradient Descent, we update the parameters using the following rule:

$$ \theta = \theta - \eta \cdot \frac{\partial L}{\partial \theta} $$

Where:

- $\theta$ represents the model parameters (like $m$ and $c$),
- $\eta$ is the learning rate (how big each step is),
- $\frac{\partial L}{\partial \theta}$ is the gradient of the loss function with respect to the parameters.

The learning rate $\eta$ controls how large the steps are during each iteration. If the steps are too small, it will take a long time to converge; if they are too large, the optimization might overshoot and not converge at all.

##### Applying Gradient Descent to Linear Regression

Let's now apply Gradient Descent to our Linear Regression example.

##### Linear Regression Example (with Gradient Descent)

We will use Gradient Descent to optimize the slope $m$ and intercept $c$ for the line that best fits the data.

```python
def f(x):
    return 2*x + 1
```

```python
# Example data (x, y)
x = np.array([1, 2, 3, 4, 5])
y = f(x)
```

```python
# Plot the data points without the line
plt.scatter(x, y, color='blue', label='Data points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data Points for Linear Regression')
plt.legend()
plt.show()
```

```python
# Define the loss function (Mean Squared Error)
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Define the gradient descent algorithm for Linear Regression
def gradient_descent(x, y, learning_rate=0.01, epochs=1000):
    m, c = 0, 0  # Initialize slope and intercept
    n = len(y)    # Number of data points
    
    # Perform Gradient Descent
    for _ in range(epochs):
        y_pred = m * x + c  # Predicted values based on current m and c
        # Calculate the gradients
        dm = (-2/n) * np.sum(x * (y - y_pred))  # Gradient of m
        dc = (-2/n) * np.sum(y - y_pred)        # Gradient of c
        # Update m and c
        m -= learning_rate * dm
        c -= learning_rate * dc
        
        # Optional: Print loss during training for debugging
        if _ % 100 == 0:
            mse = mean_squared_error(y, y_pred)
            print(f'Epoch {_}: m = {m:.4f}, c = {c:.4f}, MSE = {mse:.4f}')
    
    return m, c
```

```python
# Run Gradient Descent to find the optimal m and c
learning_rate = 0.01
epochs = 1000
m_opt, c_opt = gradient_descent(x, y, learning_rate, epochs)

print(f"Optimized slope: {m_opt}, Optimized intercept: {c_opt}")
```

```python
# Plot the data points with the optimized line
x_vals = np.linspace(0, 6, 100)
y_vals = m_opt * x_vals + c_opt
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x_vals, y_vals, color='red', label=f'y = {m_opt:.2f}x + {c_opt:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fitted Line using Gradient Descent')
plt.legend()
plt.show()
```

```python
# Calculate final MSE
y_pred = m_opt * x + c_opt
mse = mean_squared_error(y, y_pred)
print(f"Final Mean Squared Error (MSE): {mse}")
```

#### Explanation:

- We start with an initial guess for $m$ and $c$.
- In each iteration, we calculate the predictions $y_{\text{pred}}$ based on the current values of $m$ and $c$.
- We then compute the gradients $\frac{\partial L}{\partial m}$ and $\frac{\partial L}{\partial c}$, which tell us how to adjust $m$ and $c$ to reduce the error.
- After each update, we gradually move toward the values of $m$ and $c$ that minimize the Mean Squared Error (MSE).

#### Conclusion: Gradient Descent in Practice

In this section, we've extended the Linear Regression example to use Gradient Descent instead of Ordinary Least Squares. Gradient Descent is a more flexible method that can be used to optimize complex models where direct solutions like OLS are not feasible.

#### Why Gradient Descent?

- **General-purpose optimization**: Gradient Descent works for many types of models, including those that cannot be solved analytically.
- **Scalability**: In complex models with many parameters, Gradient Descent is often the only feasible way to find the optimal solution.

This is especially true when we extend this approach to Logistic Regression and more advanced machine learning models, where direct methods like OLS are not applicable.
